## Get more Modern Data

```python
from bs4 import  BeautifulSoup
import requests
```

```python
url = 'https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,"html.parser")
```

```python
elements = soup.find_all('td')
```

```python
from collections import defaultdict
```

```python
d = defaultdict(list)
```

```python
for element in elements:
    text = element.getText()
    column,value = text.split(':')
    d[column].append(value)
```

```python
modern_salaries = pd.DataFrame(d)
```

```python
modern_salaries.set_index('Rank',inplace=True,)
```

```python
modern_salaries.to_csv('2020_salaries.csv')
```

```python
modern_salaries.info()
```

```python
pd.to_numeric()
```

```python
modern_salaries.iloc[:,-3:] = modern_salaries.iloc[:,-3:].apply(lambda column:pd.to_numeric(column.str.replace(r'\D',''),errors='coerce'))
```

```python
modern_salaries.index = modern_salaries.index.astype('int')
```

```python
modern_salaries.to_csv("2020_salaries.csv")
```
